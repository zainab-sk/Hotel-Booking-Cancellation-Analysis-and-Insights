# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Loading the dataset

```python
df = pd.read_csv("hotel_bookings.csv")
```

# Exploratory Data Analysis and Data Cleaning

```python
df.head()
```

```python
df['is_canceled'] = df['is_canceled'].replace('no', 0)
```

```python
df['is_canceled'] = df['is_canceled'].replace('yes', 1)
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
```

```python
df.info()
```

```python
# describe is used for numerical columns but if we include a parametercalled include='object' so it will give summary statistics for that objects only
df.describe(include='object')
```

```python
# Check unique value or category present in each column
for col in df.describe(include='object').columns:
  print(col)
  print(df[col].unique())
  print('-'*50)
```

```python
# check for missing values in each column
df.isnull().sum()
```

```python
# Removing 'company' and 'agent' column since the missing values are very large to handle them
df.drop(['company','agent'], axis = 1, inplace = True) # Set axis = 1 to remove column and inplace= True for making change sin original dataframe
df.dropna(inplace=True)  # Remove rows with missing values
```

```python
df.isnull().sum()
```

```python
# Give summary statistics for numerical columns
df.describe()
```

```python
#Remove outlier '5400' from adr(average daily rate) can be consider as price for hotel
df = df[df['adr']<5000]
```

# Data Analysis and Visualizations

```python
# Calculating the percentage for not canceled and canceled reservations
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))
plt.title("Reservation status count")
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
# Depending on hotel find out cancelation and non cancelation counts
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel',hue = 'is_canceled',data=df,palette='Blues')
legend_labels,_= ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation status in different hotel",size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Checking if price is the factor in resulting high cancelations rate in hotel
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()

# Plot showing price for City and Resort Hotel over time
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel') # resort.index is the date column
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel') # resort.index is the date column
plt.legend(fontsize=20)
plt.show()
```

Above plot proves the hypothesis that City hotel price is comparatively cheaper than the Resort hotel

In below plot we can observe that auust month has lowest cancellations because the price is lowest in that month as observed in the plot after "Reservation status per month"

```python
# Plot shows cancelation counts over the period of months, showing which month has more cancelation and which has less
df['month']=df['reservation_status_date'].dt.month # returns month from the date column
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_= ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation status per month",size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("ADR per month", fontsize=30)

sns.barplot(x='month', y='adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index(),hue='adr')
plt.show()
```

```python
# Plot for top 10 countries having highest canceleation rates
canceled_data = df[df['is_canceled']==1]
not_canceled_data=df[df['is_canceled']==0]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct = '%.2f',labels=top_10_country.index)
plt.show()
```

### Hypothesis
The majority of clients are coming from offline travel agents to make their reservations

```python
df['market_segment'].value_counts(normalize=True)
```

```python
# Check from which hotel room booking source cancelation is happening more
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
# Checking that adr(price) is high for canceled reservations or non canceled reservations
cancelled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend(fontsize=20)
plt.show()

```

```python
# Plot for data from 2016 to 2017 september since above plot was cluttered and had missing data in 2015
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend(fontsize=20)
plt.show()
```

Average daily rate is the mostly inflencing factor that is affecting the cancellation rate
